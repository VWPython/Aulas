## Grafos
***

Grafo é uma entidade composta de duas partes:

* Vértices (nós)
* Arestas (linhas)

Os nós são as "bolinhas" (entidades que você quer modelar)

As arestas são as relações dessas entidades

![grafo1](https://user-images.githubusercontent.com/14116020/45591822-be2eaa00-b933-11e8-930f-f8c0242a2c2a.png)

O exemplo acima ilustra a relação: a cidade são paulo está ligada a cidade Buenos Aires e vice-versa com uma distância de 1680 km, ou seja, as cidades são os vértices e a distância de uma cidade a outra é a aresta

Com dois nós (vértices) é fácil de visualizar, já para milhões de nós é necessário uma boa estrutura de dados, é ai que entra as formas de representar um grafo.

Abaixo está listado as formas de representar um grafo:

* Matriz de Adjacência
* Lista de Adjacência
* Matriz de Incidência
* Lista de Incidência

***
### Qual representação utilizar?
***

A **matriz de adjacência** é boa para saber se um vértice é amigo de outro, pois basta testar matriz[v][w].

Em alguns casos, o mais barato é usar as duas representações juntas.

Em uma **lista de adjacência**, é fácil encontrar todos os vértices adjacentes a um vértice.

Em um teste de vizinhança em dois vértices, uma **matriz de adjacência** proporciona isso na hora.

A representação matricial tem grande consumo de memória para grafos grandes (muitos vértices) e com poucas arestas.

#### Ponte de um grafo

![img](https://user-images.githubusercontent.com/14116020/45592092-955de300-b93a-11e8-91c4-991b6ed98932.png)

Ponte é a aresta de corte de um grafo.

Se você retirar a aresta ponte, então o número de componentes do grafo aumentará no exemplo teremos o componente ((0, 1), (1, 2) e (2, 0)) e o componente (3, 4).

A remoção de uma ponte desconecta um grafo.

As pontes são (0, 3) e (3, 4)

Como detectar as pontes de um grafo?

Para cada aresta (u, v) faça:

* Remova a aresta (u, v) do grafo


* Verifique se o grafo permanece conectado (pode-se usar a busca em profundidade - DFS)


* Adicione a aresta (u, v) de volta ao grafo.
