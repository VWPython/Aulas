***
### Ler imagem
***

````python
from PIL import Image
from PIL import ImageEnhance
from matplotlib import pyplot as plt
````

````python
img = Image.open("recursos/image01.jpg")
````

````python
def mostrar_imagem(img):
    plt.imshow(img)
    plt.title("Imagem")
    plt.axis("off")
    plt.show()
````

````python
img.show()
mostrar_imagem(img)
````

***
### Salvar imagem
***

````python
img.save("recursos/new_image.jpg")
````

***
### Cortar imagem
***

````python
x = 100
y = 100
altura = 400
largura = 400
dimensao = (x, y, largura, altura)
````

````python
imagem_cortada = img.crop(dimensao)
````

````python
mostrar_imagem(imagem_cortada)
````

***
### Converter imagem para cinza
***

````python
# L = Escala de Cinza
imagem_cinza = img.convert("L")
````

````python
mostrar_imagem(imagem_cinza)
````

***
### Rotacionar imagem
***

````python
imagem_rotacionada = img.rotate(90)
````

````python
mostrar_imagem(imagem_rotacionada)
````

***
### Redimensionar imagem
***

````python
largura = 300
altura = 200
imagem_menor = img.resize((largura, altura))
````

````python
mostrar_imagem(imagem_menor)
````

***
### Aumentar brilho da imagem
***

````python
realce = ImageEnhance.Brightness(img)
````

````python
image_com_brilho = realce.enhance(2)
````

````python
mostrar_imagem(image_com_brilho)
````

***
### Aumentar contraste da imagem
***

````python
realce = ImageEnhance.Contrast(img)
````

````python
imagem_com_contraste = realce.enhance(2)
````

````python
mostrar_imagem(imagem_com_contraste)
````

***
### Pegar os pixels da imagem
***

````python
# (Red, Green, Blue)
coordenada = (100, 100)
pixel = img.getpixel(coordenada)
````
