### Exemplos
***

```python
# Importar o matplot e inserir o comando para gerar gráficos no notebook
import matplotlib.pyplot as matplot
%matplotlib inline
import numpy
```

***

```python
# Criar um array igualmente espaçado
x = numpy.linspace(-5, 5, 100)
y = x ** 3
```

***

```python
# Primeiro iremos criar um objeto figura
figure = matplot.figure()

# Criar os eixos laterais e posiciona-los no canvas
# (10% a esquerda, 10% a baixo, 80% da largura do canvas, 80% da altura do canvas)
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])

# Vamos plotar o gráfico nesses eixos
axes.plot(x, y)

# Setar os labels e titulos
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Título")

# Marcar o limite dos eixos
axes.set_xlim([-4, 4])
axes.set_ylim([-50, 50])
```

***

```python
# Criar um objeto figura
figure = matplot.figure()

# Criar os eixos laterais e posiciona-los no canvas
# (10% a esquerda, 10% a baixo, 80% da largura do canvas, 80% da altura do canvas)
axes1 = figure.add_axes([0.1, 0.1, 0.8, 0.8])
# (20% a esquerda, 50% a baixo, 40% da largura do canvas, 30% da altura do canvas)
axes2 = figure.add_axes([0.2, 0.5, 0.4, 0.3])

# Plotar os gráficos em x e y
axes1.plot(x, y)
axes1.set_title('Gráfico grande')

axes2.plot(y, x)
axes2.set_title('Gráfico pequeno')
```

***

```python
# Criar subplots com OO, no caso criamos 1 gráficos (1*2)
figure, axes = matplot.subplots(nrows=1, ncols=2)

# Melhorar o layout dos gráficos
matplot.tight_layout()

# Podemos verificar os arrays que contem os axes dentro de uma figura
print(figure)
print(axes)

# Vamos plotar os gráficos inserindo x e y em cada um deles
for current_axes in axes:
    current_axes.plot(x, y)
```

***

```python
# Criar subplots com OO, no caso criamos 1 gráficos (1*2)
figure, axes = matplot.subplots(nrows=1, ncols=2)

# Vamos plotar os gráficos inserindo x e y em cada um deles
axes[0].plot(x, y)
axes[0].set_title('Primeiro plot')
axes[1].plot(y, x)
axes[1].set_title('Segundo plot')
```

***

```python
# Vamos fazer o mesmo com subplots
figure, axes = matplot.subplots(nrows=2, ncols=2)

# Criar os eixos e plotar os gráficos
axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(y, x)
axes[1][1].plot(y, x)

# Melhorar o layout dos gráficos
matplot.tight_layout()
```
