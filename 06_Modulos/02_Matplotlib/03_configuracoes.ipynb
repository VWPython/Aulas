# Configurações importantes
***

**Figure size aspect ratio**: É uma tupla com a largura e altura da figura.

**GPI map**: Equivale a pixels por polegada da figura, é a escala da figura.

Ambos permitem controlar cada um dos aspectos e você pode especificar eles quando você está chamando o objeto figure()

**legend position**:

* Podemos usar coordenadas em % usando o argumento **axes.legend(loc=(0.1, 0.1))** ou por strings **axes,legend(loc='center')** ou por código **axes.legend(loc=0)**

* Para ativar as legendas os plots tem que ter um label.

|String|Code|
|------|:--:|
|'best'|0|
|'upper right'|1|
|'upper left'|2|
|'lower left'|3|
|'lower right'|4|
|'right'|5|
|'center left'|6|
|'center right'|7|
|'lower center'|8|
|'upper center'|9|
|'center'|10|

**Cores**:

* Podemos usar no plot o parametro **color** passando uma string da cor ou um hexadecimal

**Linhas**:

* **linewidth**: Largura da linha

* **alpha**: Transparencia da linha

* **linestyle**: Estilo de linha, exemplo, $-$, $--$, $-.$, $:$, $steps$

**Marcadores**:

* **marker**: Tipo de marcador de cada ponto do gráfico, por exemplo, $o$, $*$, $+$, $.$

* **markersize**: Tamanho do marcador

* **markerfacecolor**: Cor de dentro do marcador

* **markeredgewidth**: Largura das bordas do marcador

* **markeredgecolor**: Cor das bordas do marcador

***
### Exemplos
***

```python
# Importar o matplot e inserir o comando para gerar gráficos no notebook
import matplotlib.pyplot as matplot
%matplotlib inline
import numpy
```

***

```python
# Criar um array igualmente espaçado
x = numpy.linspace(-5, 5, 100)
y = x ** 3
```

***

```python
# Criando um objeto figure
figure = matplot.figure(
    figsize=(5, 2),
    dpi=100
)

# Adicionando os eixos x e y
axes = figure.add_axes([0, 0, 1, 1])

# Plotando o gráfico
axes.plot(x, y)
```

***

```python
# Declarando arrays para a confecção dos gráficos
x = numpy.linspace(-3, 3, 100)
y1 = numpy.power(x, 2)
y2 = numpy.power(x, 3)

# Criando um objeto figure
figure = matplot.figure()

# Adicionando os eixos x e y
axes = figure.add_axes([0, 0, 1, 1])

# Plotando o gráfico
axes.plot(
    x, y1,
    label='y = $x^2$',
    color='blue',
    linewidth=2,
    alpha=0.5,
    linestyle='-.'
)
axes.plot(
    x, y2,
    label='y = $x^3$',
    color='#FF8C00',
    marker='.'
)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Meu primeiro gráfico")
axes.legend(loc='lower right')
axes.grid()
```

***

```python
# Os gráficos estão armazenados no objeto figure
# Podemos salvar a figura passando a extensão
figure.savefig('graphic.png', dpi=100)
```
