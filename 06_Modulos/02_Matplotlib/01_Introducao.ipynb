# Introdução ao Matplotlib
***

O Matplotlib (http://matplotlib.org/) é um módulo para a confecção de gráficos em Python. O Matplotlib inclui diversos outros módulos, neste notebook será apresentado o matplotlib.pyplot. Este módulo funciona muito bem com o módulo Numpy e sua sintaxe é bem próxima ao MatLab. Ele será importado da forma:

```
import matplotlib.pyplot as matplot
%matplotlib inline
```

O segundo comando foi necessário para habilitar a geração de gráfico no ambiente Jupyter Notebook.

Para confeccionar gráficos, basta criar dois arrays, um para o eixo das abcsissas ou eixo X e outro para o eixo das coordenadas ou eixo Y. Então utiliza-se a função **plot()** para gerar o gráfico e a função **show()** para mostrá-lo, conforme os exemplos abaixo.

***
### Exemplos
***

````python
# Importar o matplot e inserir o comando para gerar gráficos no notebook
import matplotlib.pyplot as matplot
%matplotlib inline
import numpy
````

***

````python
# Criar um array igualmente espaçado
x = numpy.linspace(-5, 5, 100)
y = x ** 3
````

***

````python
# Vamos plotar o gráfico usando x e y, a cor e o tipo de linha
matplot.plot(x, y, 'g--')

# Criar as labels, titulosm legendas e grid
matplot.xlabel("X Label")
matplot.ylabel("Y Label")
matplot.title("Título")
matplot.grid(True)
matplot.legend(['y = $x^3$'])

# Mostrar o gráfico
matplot.show()
````

***

````python
# Declarando arrays para a confecção dos gráficos
x = numpy.linspace(-3, 3, 100)
y1 = numpy.power(x, 2)
y2 = numpy.power(x, 3)

# Gerando os gráficos
matplot.plot(x, y1)
matplot.plot(x, y2)
matplot.title('Meu primeiro gráfico')
matplot.xlabel('eixo x: x')
matplot.ylabel('eixo y: f(x)')
matplot.grid(True)
matplot.legend(['y = $x^2$', 'y = $x^3$'])

# Mostrando o gráfico
matplot.show()
````

***

````python
# Colocar multiplos gráficos no mesmo frame.
# Argumentos: Número de linhas, Número de colunas e Número de gráficos
matplot.subplot(1, 2, 1)
matplot.plot(x, y1)
matplot.title('Meu primeiro gráfico')
matplot.xlabel('eixo x: x')
matplot.ylabel('eixo y: f(x)')
matplot.grid(True)
matplot.legend(['y = $x^2$'])

matplot.subplot(1, 2, 2)
matplot.plot(x, y2)
matplot.title('Meu segundo gráfico')
matplot.xlabel('eixo x: x')
matplot.ylabel('eixo y: f(x)')
matplot.grid(True)
matplot.legend(['y = $x^3$'])

# Mostrando o gráfico
matplot.show()
````
