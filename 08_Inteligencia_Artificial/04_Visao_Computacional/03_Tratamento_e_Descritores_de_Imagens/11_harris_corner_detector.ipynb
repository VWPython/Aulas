## Harris Corner Detector
***

Este algoritmo específico foi desenvolvido para identificar o canto interno de uma imagem. Os cantos são regiões dentro de uma imagem em que há grandes variações na intensidade do gradiente em todas as direções.

O Harris Corner Detector foi introduzido em 1988 por Harris e Stephens em seu documento, A Combined Corner and Edge Detector. Este detector é um dos detectores de canto mais comuns que você encontrará no mundo da visão computacional. É bastante rápido (não tão rápido como o detector de ponto-chave FAST), mas é mais preciso ao marcar as regiões como cantos.

***

```python
# Imports
import numpy as np
from matplotlib import pyplot as plt
import cv2


def show(img, title=""):
    """
    Redimensionar a imagem e mostrar no codigo.
    """

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.rcParams['figure.figsize'] = (50, 50)
    plt.title(title, fontdict={"fontsize": 100})
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
# Leitura da image e conversão para Grayscale
img = cv2.imread('imagens/box.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Conversão para float32
gray = np.float32(gray)
```

cv2.cornerHarris(image, blockSize, ksize, k)

Esta função recebe quatro argumentos.

* **img** - imagem a ser analisada, deve estar em escala de cinza e com valores float32.
* **blockSize** - tamanho das janelas consideradas para a detecção de canto
* **ksize** - parâmetro para a derivada de Sobel
* **k** - parâmetro livre para a equação de Harris.

```python
# Cantos
corners = cv2.cornerHarris(gray, 4, 5, 0.04)
```

```python
# O resultado é dilatado para marcar os cantos
corners = cv2.dilate(corners, None)
```

```python
# Limiar para um valor ideal; pode variar dependendo da imagem.
img[corners > 0.01 * corners.max()] = [0,0,0]
```

```python
show(img)
```

```python
# Canto inferior direito
img = cv2.imread('imagens/box.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
corners = cv2.cornerHarris(gray, 9, 5, 0.04)
corners = cv2.dilate(corners, None)
img[corners > 0.01 * corners.max()] = [0,0,0]
show(img)
```

```python
# Canto superior esquerdo
img = cv2.imread('imagens/box.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
corners = cv2.cornerHarris(gray, 14, 5, 0.04)
corners = cv2.dilate(corners, None)
img[corners > 0.01 * corners.max()] = [0,0,0]
show(img)
```
