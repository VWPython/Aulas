## Introdução
***

Neste modulo de visão computacional será apresentado as 4 principais áreas da Visão Computacional: Classificação, Detecção, Localização e Segmentação de Imagens.

***
### O que é visão computacional
***

Visão computacional é o estudo da extração de informações de uma imagem; mais especificamente, é a construção de descrições explícitas e claras dos objetos em uma imagem.

A visão computacional nada mais é que a tentativa de reproduzir a visão humana, ou seja, é o processo de modelagem e replicação da visão humana usando software e hardware.

Visão computacional e reconhecimento de imagem são termos frequentemente usados como sinônimos, mas o primeiro abrange mais do que apenas analisar imagens.

A visão computacional é uma área que envolve várias outras áreas como: Inteligência Artificial, Machine Learning, Matemática, Neurobiologia, Tratamento de imagens, Física, Processamento de sínais, robotica.

### Tipos de visão computacional
***

#### Classificação:

![img](https://user-images.githubusercontent.com/14116020/55771496-ad27c080-5a5e-11e9-9cda-d6e5a59c6627.png)

É a forma mais simples de visão computacional, ou seja, eu entrego uma imagem ao modelo e ele vai apenas classificar a image, por exemplo, se é uma imagem de um gato, cachorro, carro e etc...

Para realizar a classificação o modelo tem que ser treinado previamente, ele aprende os atributos de um objeto na imagem e sempre que for entregue uma nova imagem ele consegue saber se existe ou não o objeto aprendido.

#### Localização

![img](https://user-images.githubusercontent.com/14116020/55771513-bd3fa000-5a5e-11e9-889f-28f018ed789d.png)

É utilizado em conjunto com a classificação, é utilizando para localizar a posição desse objeto detectado.

#### Detecção

![img](https://user-images.githubusercontent.com/14116020/55771555-ec561180-5a5e-11e9-93b6-47b391b54d75.png)

Em conjunto aos outros dois ou não, é usado para detectar vários objetos em uma imagem.

#### Segmentação

![img](https://user-images.githubusercontent.com/14116020/55771632-3a6b1500-5a5f-11e9-94f7-71f93139d191.png)

Para completar temos a segmentação, ou seja, é traçar exatamente o contorno dos objetos detectados dentro da imagem. Um exemplo, é na área medica que preciso saber exatamente uma área no diagnostico do pulmão onde se localiza uma doença, por exemplo, um tumor.

### Computer Vision vs Machine Vision vs Processamento de Imagem
***

#### Computer Vision

É o termo mais abrangente, ou seja, dentro da visão computacional temos a visão de máquina e o processamento de imagem, além da morfologia de imagens e análise de imagens.

#### Machine Vision

É a aplicação de visão computacional para automação, também engloba processamento de imagem, morfologia e análise de imagens. Por exemplo, um carro autonomo ta nessa categoria.

#### Processamento de Imagem

Fazer modificações ou melhoria em imagens está nessa sub-área da visão computacional.

### Visão Humana vs Visão Computacional
***

#### Visão Humana

![img](https://user-images.githubusercontent.com/14116020/55771958-91251e80-5a60-11e9-92a3-3a9532829d35.png)

**1)**  Ao olhar essa imagem você está recebendo uma entrada de luz, essa luz reflete através dos objetos seguindo em linha reta para o olho.

**2)** Com isso a luz passa através da **cornea**, então ela entra e acessa a **pupila** através do **cristalino**.

**3)** A **cornea** e o **cristalino** se ajustam a luz (refratar) para que então a **retina** possa focar a imagem.

**4)** O **fotorreceptores** presentes na **retina** vão converter a luz em **impulsos eletricos**.

**5)** Os impulsos eletricos atravessam o **nervo optico** até o cérebro

**6)** O **cérebro** processa os sínais para criar a imagem que vc está visualizando nesse momento

Isso tudo acontece em milisegundos.

#### Visão Computacional

![img](https://user-images.githubusercontent.com/14116020/55772170-7bfcbf80-5a61-11e9-955a-bd45ff2aa70d.png)

**1)** Tudo começa com uma **análise qualitativa** onde nós temos a aquisição ou digitalização da imagem, transformando-a em pixels.

**2)** Esses pixels serão passados por um processo de restauração e realce, fazendo com que a imagem fique melhor e mais nítida ao nosso objetivo.

**3)** Em sequências passamos pelo processo de segmentação desses pixels, ou seja, dentro da imagem buscar apenas o pixels que me interessa, formando grupos de pixels.

**4)** Com esses grupos iremos extrair atributos e características formando um conjunto de dados.

**5)** Com esse conjunto de dados (atributos e características) podemos realizar classificações e reconhecimento nessa imagem, gerando vários grupos de dados.

**6)** Esses grupos de dados serão usado para **análises quantitativas** para a tomada de decisões.

A visão computaciona tem algumas particularidades em relação a visão humana que pode atrapalhar no processo, por exemplo, a resolução e qualidade da imagem, quantização da imagem, ou seja, quantidade de cores que ela apresenta, concentração de pixels (resolução espacial).

Um **pixel** é caracterizado pelo valor de tonalidade de cor e pela sua localização na imagem.

Esse quadrado é uma representação matricial de uma região da imagem.

### Ferramentas e Bibliotecas
***

* **Linguagem de Programação**: O Python é a linguagem difundidas para visão computacional

* **OpenCV**: Principal biblioteca para visão computacional.

* **SimpleCV**: Outra biblioteca para visão computacional mais simples.

* **ROS**: Robotic Operating System, tem uma série de funcionalidades para visão computacional aplicadas a robos em C++.

* **FFmpeg**: É um programa para extrair informações de videos

* **CUDA**: É uma ferramenta para programação paralela em GPUs, dependendo da tarefa de visão computacional pode ser necessário utilizar essa plataforma.
