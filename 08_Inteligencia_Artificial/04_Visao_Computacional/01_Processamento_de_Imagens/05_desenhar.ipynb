## Desenhar em imagens
***

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def mostrar_imagem(img):
    imagem_colorida = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(imagem_colorida)
    plt.title("Imagem")
    plt.axis("off")
    plt.show()
```

```python
# Inicializando nossa tela com um diametro de 300x300 e 3 canais (RGB) com um fundo preto
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
# Desenhando uma linha verde no canto superior esquerdo da nossa tela
# para o canto inferior direito
verde = (0, 255, 0)
canto_superior_esquerdo = (0, 0)
canto_inferior_direito = (300, 300)
linha = cv.line(
    img=canvas,
    pt1=canto_superior_esquerdo,
    pt2=canto_inferior_direito,
    color=verde
)
mostrar_imagem(canvas)
```

```python
# Desenhando uma linha vermelha de 3 pixels de espessura do canto superior direito
# para o canto inferior esquerdo
canto_superior_direito = (300, 0)
canto_inferior_esquerdo = (0, 300)
linha = cv.line(
    img=canvas,
    pt1=canto_superior_direito,
    pt2=canto_inferior_esquerdo,
    color=(0, 0, 255),
    thickness=3
)
mostrar_imagem(canvas)
```

```python
# Desenhando um quadrado verde de 50x50 pixels, começando em 10x10 e termina em 60x60
preenchido = -1
retangulo = cv.rectangle(
    img=canvas,
    pt1=(10, 10),
    pt2=(60, 60),
    color=(0, 255, 0),
    thickness=preenchido
)
mostrar_imagem(canvas)
```

```python
# Desenhando um retangulo vermelho com 5 pixels de espessura
retangulo = cv.rectangle(
    img=canvas,
    pt1=(50, 200),
    pt2=(200, 225),
    color=(0, 0, 255),
    thickness=5
)
mostrar_imagem(canvas)
```

```python
# Desenhando um retangulo vermelho com 5 pixels de espessura
retangulo = cv.rectangle(
    img=canvas,
    pt1=(200, 50),
    pt2=(225, 125),
    color=(255, 0, 0),
    thickness=-1
)
mostrar_imagem(canvas)
```

```python
# Desenhando um circulo branco no centro da tela com raios crescentes de 25 pixels para 150 pixels
altura = canvas.shape[0]
largura = canvas.shape[1]
(centerX, centerY) = (largura // 2, altura // 2)
for raio in range(0, 175, 25):
    cv.circle(
        img=canvas,
        center=(centerX, centerY),
        radius=raio,
        color=(255, 255, 255)
    )
    
mostrar_imagem(canvas)
```
