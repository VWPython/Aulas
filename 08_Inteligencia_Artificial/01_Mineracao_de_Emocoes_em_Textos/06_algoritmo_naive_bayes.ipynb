## Algorítmo Naive Bayes
***

Esse algoritmo tem uma abordagem totalmente probabilística (Teorema de Bayes). Esse teorema é muito utilizado por estatísticos para a tomada de decisão.

Exemplos que podem ser feitos com esse algoritmo:

* **Filtros de Spam**: Quando você recebe um email ele pode ser classificado em 2 categorias, SPAM ou NÂO SPAM
* **Mineração de Emoções**: Computação afetiva, dado um texto ou voz identificar qual a emoção que está naquela entrada.
* **Separação de documentos**: Identificar e categorizar documentos e notícias.
* **Mineração de textos**: Algoritmo muito utilizado para classificar e organizar textos.

O algorítmo naive bayes vai ler a base de dados original e criar uma tabela de probabilidades.

Podemos concluir com essa tabela, por exemplo, que 1 registro com boa história de créditos entre 5 existentes tem risco alto de emprestimo.

Aplicando um novo registro nessa tabela treinada pelo algorítmo de naive bayer irá retorna ser o risco de emprestimo é baixo, médio ou alto de acordo com sua probabilidade de ocorrencia.

Vamos supor que foi inserido um novo registro em que temos como história de créditos **boa**, dívida **alta**, garantia **nenhuma** e sua renda é **maior que 35 mil** por ano. Qual a probabilidade dessa pessoa ter **risco alto**, **moderado** ou **baixo**?

* **Probabilidade(Alto)** = $\frac{6}{14} \times \frac{1}{5} \times \frac{4}{7} \times \frac{6}{11} \times \frac{1}{7} = 0,0038 = \frac{0,0038}{0,0158}\times100 = 24,05\%$


* **Probabilidade(Moderado)** = $\frac{3}{14} \times \frac{1}{5} \times \frac{1}{7} \times \frac{2}{11} \times \frac{1}{7} = 0,0001 = \frac{0,0001}{0,0158}\times100 = 0,63\%$


* **Probabilidade(Baixo)** = $\frac{5}{14} \times \frac{3}{5} \times \frac{2}{7} \times \frac{3}{11} \times \frac{5}{7} = 0,0119 = \frac{0,0119}{0,0158}\times100 = 75,31\%$

$0,0038 + 0,0001 + 0,0119 = 0,0158$ esse número equivale a $100\%$

Com isso a probabilidade do risco ser baixo é maior do que a de ser alto ou moderado, com isso podemos concluir que o risco para esse registro é baixo
