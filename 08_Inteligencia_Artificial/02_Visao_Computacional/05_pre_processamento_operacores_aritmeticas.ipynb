## Pré Processamento: Operações Aritméticas
***

Com as operações aritméticas podemos adicionar ou remover conteúdos nas imagens

***
### Operação de Adição
***

Somar os valores dos pixels de duas imagens

```py
add(img1, img2):
    Entrada = Imagem1 e Imagem2
    Saída = Matriz com os valores dos pixels somados
```

````python
import cv2 as cv
from matplotlib import pyplot as plt
````

````python
img1 = cv.imread("./Imagens/superman.png")
img2 = cv.imread("./Imagens/batman.png")
````

````python
img_somada = cv.add(img1, img2)
````

````python
# Mostrar a imagem
plt.imshow(img_somada)
plt.title("Imagem Somada")
plt.show()
````

***
### Operação de Subtração
***

Diferenças entre duas imagens, uma fica sobresposta a outra

```py
subtract(img1, img2):
    Entradas: Imagem 01 e Imagem 02
    Saidas: Matriz com os valores dos pixels subtraídos
```

````python
img_subtraida = cv.subtract(img1, img2)
````

````python
# Mostrar a imagem
plt.imshow(img_subtraida)
plt.title("Imagem Subtraida")
plt.show()
````

***
### Operação de Mistura
***

Mesclar, com perda de dados, as informações de duas imagens em apenas uma

```py
addWeighted(src1, alpha, src2, beta, gamma):
    src1 = Matriz referente à primeira imagem
    alpha = Intensidade da primeira imagem
    src2 = Matriz referente à segunda imagem
    beta = Intensidade da segunda imagem
    gamma = Valor escalar adicionado a cada soma
    
    return Matriz de uma imagem mesclada
```

alpha e beta varia de 0.0 (transparente) até 1.0 (densidade máxima)

gamma realiza ajustes o brilho da imagem

````python
img_misturada = cv.addWeighted(img1, 0.8, img2, 1, 0)
````

````python
plt.imshow(img_misturada)
plt.title("Imagem Mesclada")
plt.show()
````
