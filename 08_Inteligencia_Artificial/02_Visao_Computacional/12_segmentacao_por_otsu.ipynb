## Segmentação por OTSU
***

![img](https://user-images.githubusercontent.com/14116020/54875402-3b544380-4ddd-11e9-99af-bffad353951f.png)

Na segmentação binaria o usuário deve informar o valor da limiar, considere uma imagem bimodal, ou seja, uma imagem cujo histograma tem 2 picos, como mostrado na imagem central, podemos ter um valor limite no meio desses dois picos e é isso que a segmentação por OTSU faz.

A segmentação por OTSU calcula automaticamente o valor limite do histograma da imagem para uma imagem bimodal, ou seja, o usuário não precisa adivinhar um limear o algoritmo de OTSU calcula automaticamente.

A primeira imagem a seguir, representa uma imagem de um coração com limear de 40% e a segunda é o histograma da imagem do coração, já a ultima é essa mesma imagem com limear calculado pelo algoritmo de OTSU por meio do histograma do coração.

![img](https://user-images.githubusercontent.com/14116020/54875390-f6c8a800-4ddc-11e9-9a99-29397734c44b.png)

```py
def threshold(img, limiar, val_int, método):
    img = Imagem com a qual queremos trabalhar em tons de cinza
    limiar = valor estipulado pelo usuário (tentativa e erro)
    val_int = Valor de intensidade dos pixels com valores maiores que o limiar
    método = THRESH_BINARY - Objeto de interesse em cor preta
             THREAH_BINARY_INV - Objeto de interesse em cor branca
             
    return Matriz referente a imagem binarizada
```

Método = THRESH_BINARY + THRESH_OTSU

````python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
````

````python
img = cv.imread("./Imagens/cafe.jpg", cv.IMREAD_GRAYSCALE)
````

````python
metodo = cv.THRESH_BINARY_INV + cv.THRESH_OTSU
````

````python
limiar, img_binario = cv.threshold(img, 0, 255, metodo)
````

````python
img_gray = cv.cvtColor(img, cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
plt.title("Imagem Original")
plt.show()
````

````python
plt.hist(img.ravel(), 256, [0, 256])
plt.show()
````

````python
img_gray = cv.cvtColor(img_binario, cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
plt.title("Imagem Binaria segmentada")
plt.show()
````
