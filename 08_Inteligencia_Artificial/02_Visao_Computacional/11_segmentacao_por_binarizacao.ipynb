## Segmentação por Binarização
***

![img](https://user-images.githubusercontent.com/14116020/54875159-f679de00-4dd7-11e9-9c4b-88fa6dc4dbd4.png)

É uma das principais etapas da visão computacional

Essa etapa consiste em separar somento o objeto de interesse em uma nova imagem, excluindo o segundo plano da imagem original.

É o primeiro passo para que possamos extrair caracteristicas de um objeto, e através dela é possível obter informações e detalhes que torna possível classifica-lo.

![img](https://user-images.githubusercontent.com/14116020/54875195-e7476000-4dd8-11e9-8b68-8f47ac0fce28.png)

A segmentação por binarização também é conhecida como aplicação de linear de intensidade. Pixels representados por valores maiores que o limiar são estabelecidos como os objetos de interesse (brancos ou pretos).

```py
def threshold(img, limiar, val_int, método):
    img = Imagem com a qual queremos trabalhar em tons de cinza
    limiar = valor estipulado pelo usuário (tentativa e erro)
    val_int = Valor de intensidade dos pixels com valores maiores que o limiar
    método = THRESH_BINARY - Objeto de interesse em cor preta
             THREAH_BINARY_INV - Objeto de interesse em cor branca
             
    return Matriz referente a imagem binarizada
```

````python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
````

````python
img = cv.imread("./Imagens/cafe.jpg")
````

````python
metodo = cv.THRESH_BINARY_INV
````

````python
limiar, img_binario = cv.threshold(img, 200, 255, metodo)
````

````python
plt.imshow(img)
plt.title("Imagem Original")
plt.show()
````

````python
plt.imshow(img_binario)
plt.title("Imagem Binaria segmentada")
plt.show()
````

### Binarização adaptativa
***

![img](https://user-images.githubusercontent.com/14116020/54875265-64bfa000-4dda-11e9-861c-5de869f0e6f7.png)

Calcula diferentes valores de limiar para cada região da imagem. Muito boa quando o objeto tem boa iluminação.

Obtem o melhor resultado considerando o contraste

```py
def adaptiveThreshold(src, maxValue, adaptativeMethod, thresholdType, blockSize, C):
    src = Matriz referente à imagem.
    maxValue = Valor de intensidade máxima do pixel.
    adaptiveMethod = ADAPTIVE_THRESH_MEAN_C
                     ADAPTIVE_THRESH_GAUSSIAN_C
    thresholdType = THRESH_BINARY - Objeto de interesse em cor preta
                    THRESH_BINARY_INV - Objeto de interesse em cor branca
    blockSize = Tamanho da máscara
    C = Constante de subtração da média ou da média ponderada.
    
    return Matriz referente a imagem binarizada adaptada.
```

````python
img = cv.imread("./Imagens/olho.png", cv.IMREAD_GRAYSCALE)
````

````python
img_gauss = cv.medianBlur(img, 7)
````

````python
img_binarizada_mean = cv.adaptiveThreshold(img_gauss, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 5)
img_binarizada_gauss = cv.adaptiveThreshold(img_gauss, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
````

````python
img_gray = cv.cvtColor(img, cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
plt.title("Imagem Original")
plt.show()
````

````python
img_gray = cv.cvtColor(img_binarizada_mean, cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
plt.title("Imagem Media")
plt.show()
````

````python
img_gray = cv.cvtColor(img_binarizada_gauss, cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
plt.title("Imagem Gaussiana")
plt.show()
````
