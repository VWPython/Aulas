## Classificador Cascade
***

A visão computacional tem como objetivo simular a capacidade do olho humano e é uma sub área da Inteligência Artificial.

Para treinar um algoritmo de detecção de objetos ou faces, temos que ter um conjunto enorme de imagens positivas e negativas, por exemplo, para identificar faces, o conjunto positivo é só de rosto de pessoas e o negativo é tudo que não é rosto de pessoas como paisagens, carros, animais e etc...

Com isso iremos fazer o treinamento das imagens usando o algoritmo chamado **AdaBoost**, com ele você seleciona o melhor conjunto de caracteristicas para identificar os objetos que você quer.

O resultado desse treinamento é um arquivo XML, ou seja, cada arquivo XML vai conter um conjunto de caracteristicas para identificar algo na imagem, por exemplo, um olho, um objeto, um sorriso e etc...

Com o algoritmo pronto, quando você tem uma nova imagem e você quer fazer a detecção das faces dessa nova imagem, você vai aplicar o conjunto de características acima nessa imagem pixel por pixel.

***
### Haar Cascade
***

O Haar é cada caracteristicas ou quadradinhos usados em cada pixel da imagem.

![img1](https://user-images.githubusercontent.com/14116020/55293279-56214c00-53cb-11e9-8441-c2583e4915d5.png)

O HaarCascade é o conjunto das seguintes caracteristicas: Bordas, Linhas e Centros, ou seja, uma combinação dessas caracteristicas se forma um objeto a ser identificado na imagem ou um classificador.

É um padrão retangular que vai passar de N em N pixels da imagem aplicando o algoritmo. Existem diferenças na intensidade das regiões retangulares da imagem. Essa intensidade será diferenciada pelas cores branco e preto no algoritmo.

A caracteristica é a diferença entre a soma dos pixels brancos e os pixels pretos, o processamento de imagem é algo bem custoso, ela pode gerar mais de 160.000 combinações em uma imagem 24x24.

***
### Algoritmo AdaBoost
***

Esse algoritmo vai remover as características não necessárias e tem por objetivo combinar vários classificadores fracos, ou seja, cada quadradinho, em um classificador forte (combinação de vários quadradinhos).

Por exemplo, de 160.000 caracteristicas ele vai escolher, por exemplo, somente 6.000.

***
### Cascade
***

Cascata é o processo que será utilizado para realizar a classificação.

1. Vai deslizar pela imagem
2. Computar a média dos valores dos pixels na área branca e preta
3. Se a diferença entre as áreas é abaixo de um limiar, a característica coincide (match)

Esse método é baseado em aprendizagem supervisionada, ou seja, ele indica quem é quem, por meio de imagens positivas e negativas.

Dado uma nova imagem, ela passa por N classificadores, se existir a caracteristica ele vai para a próxima caracteristicas, caso não exista ele retorna False e não a detecção daquela caracteristica. Ou seja, no final dessa cascata de caracteristicas ele vai detectar o objeto desenhando um retangulo sobre ele.

### Detecção vs Reconhecimento
***

**Detecção**: É mostrar se existe ou não algum objeto.

**Reconhecimento**: Você além de detectar o objeto diz o que é ele ou quem é ele. Muito utilizado pelo facebook.

#### Aplicações de Detecção

* Contar quantas pessoas estão no ambiente
* Alarmes que dispara quando detecta alguma pessoa
* Controle de tráfego em rodovias
* Obter o tamanho de um objeto
* Detecção de sorrisos em câmera e ajuste de foco

#### Aplicações de Reconhecimento

* Desbloqueio do celular
* Sistemas de segurança
* Alarmes com reconhecimento do dono
* Validação em cursos on-line
* Identificação de bandidos

**OpenCV**: Open Source Computer Vision, é a biblioteca mais popular para detecção de objetos e processamento digital de imagens. Escrita em C/C++ porém podem ser usadas em python e Java que chama o C/C++.
