## Pré Processamento: Filtros de Suavização
***

São matrizes que percorrem toda a imagem alterando os valores dos pixels com objetivo de corrigir, realçar ou suavizar determinadas regiões da imagem.

Esse procedimento de percorrer a imagem com a mascara em cima é chamada de **convolução**.

![img](https://user-images.githubusercontent.com/14116020/54874119-b3ae0b00-4dc3-11e9-9cdf-0ff75c179bf2.png)

A matriz a esquerda representa uma imagem em tons de cinza. A matriz do meio é a mascara que será aplicada a imagem, e a terceira imagem é o resultado da convolução. Na qual o pixel alvo localizado na linha 3 e coluna 3 foi alterado de 11 para 2.

Na operação de convolução em cada interação do processo a mascara sobrepõem 1 pixo da imagem submetida, em cada uma dessa etapas é realiada o produto ponto a pontos dos elementos das regiões sobrepostas com o elemento da mascara.

Os filtros podem ser classificados como lineares e não lineares, a diferença entre eles é o tipo de operação realizado entre a mascara e o pixel alvo da imagem.

Os **filtros lineares** são os mais usados, eles usam a mascara e realizam somas ponderadas dos pixels ao longo da imagem. Eles são capazes de suavizar, realçar detalhes e minimizar efeitos de ruidos sem alterar o nível de cinza da imagem. Eles podem ser classificados como **passa-baixas** e **passa-altas** e referem-se ao tipo de frequência que será filtrado. Um filtro passa-baixa atenua as altas frequências, ou seja, suavizam a imagem atenuando regiões que representam bordas ou contornos. Já as passa-altas atenua as baixas frequências, realçando as regiões de bordas ou contornos.

Os **filtros não lineares** usam a mascara e realizam somas não ponderadas dos pixels ao longo da imagem. Na prática eles podem alterar o nível médio de cinza da imagem, por exemplo, destacar bordas, linhas ou manchas.

````python
import cv2 as cv
from matplotlib import pyplot as plt
````
