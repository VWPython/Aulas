## Gradiente morfológico
***

As operações morfológicas tem três tipos:

* **Gradiente Morfológico**: É a diferença entre a dilatação e a erosão de uma imagem, esse resultado é uma imagem que representa uma borda do objeto de interesse.


![img](https://user-images.githubusercontent.com/14116020/54875067-5b343900-4dd6-11e9-83bb-b6530bc74c15.png)

* **TopHat**: Diferença entre a imagem original e a abertura da imagem. É usada para realçar objetos brilhantes em fundos escuros além de corriger avaliação de luminosidade em imagens possibilitando que objetos representado regiões mais escuras sejam realçadas.

![img](https://user-images.githubusercontent.com/14116020/54875070-70a96300-4dd6-11e9-821d-46161b3a2bd1.png)

* **BlackHat**: Diferença entre o fechamento da imagem de entrada e a imagem de entrada. É oposto da TopHat.

```py
def morphologyEx(img, oper, elem_estr):

    oper = Operação que será realizada

    return Imagem nova
```

````python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
````

````python
img1 = cv.imread("./Imagens/moeda.png")
img2 = cv.imread("./Imagens/esqueleto.tif")
````

````python
elemento_estruturante1 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))
elemento_estruturante2 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (21, 21))
````

````python
img_processada1 = cv.morphologyEx(img1, cv.MORPH_GRADIENT, elemento_estruturante1)
img_processada2 = cv.morphologyEx(img2, cv.MORPH_TOPHAT, elemento_estruturante2)
img_processada3 = cv.morphologyEx(img2, cv.MORPH_BLACKHAT, elemento_estruturante2)
````

````python
plt.imshow(img1)
plt.title("Imagem Original")
plt.show()
````

````python
plt.imshow(img_processada1)
plt.title("Imagem Processada com gradiente morfologico")
plt.show()
````

````python
plt.imshow(img2)
plt.title("Imagem Original")
plt.show()
````

````python
plt.imshow(img_processada2)
plt.title("Imagem Processada com tophat")
plt.show()
````

````python
plt.imshow(img_processada3)
plt.title("Imagem Processada com blackhat")
plt.show()
````
