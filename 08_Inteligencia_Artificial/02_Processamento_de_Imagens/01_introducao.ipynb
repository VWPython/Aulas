## Visão Computacional
***

A visão computacional consiste em captar imagens, melhorá-las, separar as regiões ou objetos de interesse de uma cena, extrair várias informações e finalmente relacionar as imagens com outras vistas previamente.

Visão computacional é capaz de realizar análises com precisão e velocidade superior ao olho humano.

Áreas que a visão computacional se aplica:

* Controle de qualidade
* Astronomia - descobertas de novos planetas
* Análise biológica em níveis microscópio
* Veículos autônomos

Funcionamento:

![img](https://user-images.githubusercontent.com/14116020/54872563-8736c680-4da4-11e9-83d2-0d09d95966ac.png)

Ao captar uma imagem ou vídeo com os olhos o nosso cérebro processa a imagem e a classifica. Na visão computacional é parecida, a imagem ou vídeo é captado por uma webcam por exemplo, processada por um computador para ai depois ser interpretada.

![img](https://user-images.githubusercontent.com/14116020/54872577-ccf38f00-4da4-11e9-9095-259f6c458beb.png)

A visão computacional abrange várias áreas como: Processamento de Imagem, Aprendizado de Máquina, Recuperação de Informação e Sistemas cognitivos e Neurociência.

### Uma breve descrição da visão humana
***

O orgão responsável pelo sentido da visão humana é o olho.

![img](https://user-images.githubusercontent.com/14116020/54872680-830ba880-4da6-11e9-923d-1b7a7aae14c5.png)

A luz refletida em um objeto passa pela córnea e a pupila, estruturas que regulam a entrada de luz nos olhos por variação do seu diâmentro.

Em seguida passa pelo cristalino, estrutura que funciona como uma lente biconvexa que focaliza a luz na retina.

A retina, por sua vez, é uma camada de tecido nervoso no fundo dos olhos contituída por dois tipos de fotorreceptores.

* **Bastonetes**: Células com alta sensibilidade à iluminação e que não reconhecem diferenças de cor.

* **Cones**: Células que necessitam de maiores intensidades de luz, possuem maior acuidade e reconhecem cores.

O quiasma óptico recebe os impulsos pelos fotorreceptores pelo nervo óptico. Essa estrutura divide partes dos impulsos de um olho para o hemisfério do mesmo lado e parte dos impulsos para o hemisfério do lado oposto.

Por fim as informações chegam em uma parte denominada córtex visual.

### Sistema de visão computacional
***

Um sistema de visão computacional é dividido em várias fáses:

* Aquisição
* Processamento da Imagem
* Segmentação
* Análise da Imagem
* Reconhecimento de padrões

#### Aquisição

Fase responsável pela captação da imagem, ou seja, é a parte que simula nossos olhos, normalmente utilizando lentes como scanners, máquinas fotográficas, filmadoras, webcam e etc...

![img](https://user-images.githubusercontent.com/14116020/54872758-5a84ae00-4da8-11e9-9768-777b79a601a4.png)

#### Processamento de imagens

Fase responsável por melhorar a imagem. Retirar ruídos, suavisar a imagem e etc.

Nesta fase pode estar incluido o pré-processamento como cortes, rotação da imagem e etc...

![img](https://user-images.githubusercontent.com/14116020/54872720-62901e00-4da7-11e9-8d38-00eff88816b4.png)

#### Segmentação

Fase responsável por particionar a imagem em regiões de interesse, na parte da esquerda da imagem é a tomografia original já a parte da direita é a tomografia segmentada com as partes que queremos analisar.

![img](https://user-images.githubusercontent.com/14116020/54872728-8d7a7200-4da7-11e9-91ae-7d0a76ccb320.png)

#### Análise da imagem

Fase responsável por obter um conjunto de características ou representações númericas do objeto de interesse.

Por exemplo a imagem abaixo a esquerda é a forma como nós vemos a imagem, já na direita é a forma como o PC vê a imagem, ou seja, em forma de pixels com valores númericos representado-as.

![img](https://user-images.githubusercontent.com/14116020/54872742-f4982680-4da7-11e9-9143-2be229d65a36.png)

Assim podemos fazer uma extração de caracteristicas por causa desses valores que os pixels tem.

#### Reconhecimento de padrões

Responsável por classificar ou agrupar as imagens com base em seus conjuntos de características.

![img](https://user-images.githubusercontent.com/14116020/54872749-27421f00-4da8-11e9-9d3e-48cb8ec13c1b.png)
