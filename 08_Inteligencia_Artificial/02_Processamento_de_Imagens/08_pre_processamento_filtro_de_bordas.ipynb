## Pré Processamento: Filtro de bordas
***

Além do tratamento de ruidos os filtros também podem ajudar na extração de caracteristicas como detecção ou realce de bordas, linhas e contornos da imagem.

Como os filtros laplacianos e sobel são sensiveis a ruídos, deve-se usar filtros passa-baixas na imagem antes de utiliza-los.

***
### Filtro Sobel
***

![img](https://user-images.githubusercontent.com/14116020/54874631-f96fd100-4dcd-11e9-9d93-f95005f4d9bd.png)

Filtro espacial não linear usado para realçar bordas verticais e horizontais

```py
def Sobel(img, var_arm, h, v, máscara):
    img = Imagem com a qual queremos trabalhar
    var_arm = Valor que representa o pixel
    h = Realce horizontal
    v = Realce vertical
    máscara = Dimensão da máscara que será aplicada.
    
    return Matriz referente a imagem com o realce do filtro Sobel
```

Se o h for 1 e v for 0 o realce será só horizontal.

A máscara será valores impares inteiros de 1 a 7

````python
import cv2 as cv
from matplotlib import pyplot as plt
````

````python
img_original = cv.imread("./Imagens/predio.jpg")
````

````python
sobel_x = cv.Sobel(img_original, cv.CV_8U, 1, 0, ksize=5)
sobel_y = cv.Sobel(img_original, cv.CV_8U, 0, 1, ksize=5)
````

````python
plt.imshow(img_original)
plt.title("Imagem Original")
plt.show()
````

````python
plt.imshow(sobel_x)
plt.title("Imagem com realce nas bordas horizontais")
plt.show()
````

````python
plt.imshow(sobel_y)
plt.title("Imagem com realce nas bordas verticais")
plt.show()
````

### Filtro Laplaciano
***

![img](https://user-images.githubusercontent.com/14116020/54874719-a1d26500-4dcf-11e9-9fd1-d3c5bf0a0868.png)

Filtro espacial não linear que possui uma máscara de ordem 3.

```py
def Laplacian(img, var_arm):
    img = Imagem com a qual queremos trabalhar
    var_arm = Valor que representa o pixel CV_8U (pixels de 0 a 255 e armazenado em 8 bites)
    
    return Matriz referente a imagem com o realce do filtro laplaciano
```

````python
img_laplace = cv.Laplacian(img_original, cv.CV_8U)
````

````python
plt.imshow(img_original)
plt.title("Imagem Original")
plt.show()
````

````python
plt.imshow(img_laplace)
plt.title("Imagem com realce laplaciano")
plt.show()
````
